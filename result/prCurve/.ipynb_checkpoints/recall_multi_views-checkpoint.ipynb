```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# draw the picture
def draw(baseline_merge, topic_day, topic_location, context_day, context_location, filename):

    plt.figure(figsize=(8, 7))

    baseline_merge,   = plt.plot(baseline_merge[0], baseline_merge[1],     'k', linewidth=4)
    topic_day,        = plt.plot(topic_day[0], topic_day[1],     'r--', linewidth=4)
    topic_location,   = plt.plot(topic_location[0], topic_location[1], 'g--', linewidth=4)
    context_day,      = plt.plot(context_day[0], context_day[1],   'r', linewidth=4)
    context_location, = plt.plot(context_location[0], context_location[1], 'g', linewidth=4)
    
    # set X axis
    plt.xlim( [0, 5200] )
    plt.xticks( np.linspace(0, 5000, 6), size=15)
    plt.xlabel("Top U", fontsize=20)
    
    # set Y axis
    plt.ylim( [0.2, 1.05] )
    plt.yticks( np.linspace(0.2, 1, 9), size=15)
    plt.ylabel("Recall",    fontsize=20)
    
    # set figure information
    # plt.title("", fontsize="x-large")
    plt.legend([baseline_merge, topic_day, topic_location, context_day, context_location], ("Baseline-Merge", "Topic-Day", 
    "Topic-Location", "Context-Day-Location", "Context-Location-Time"), loc="lower right", numpoints=1, fontsize=18)
    plt.grid(True)

    # draw the chart
    plt.savefig("./" + filename + ".pdf", format="pdf")
    plt.show()
```
